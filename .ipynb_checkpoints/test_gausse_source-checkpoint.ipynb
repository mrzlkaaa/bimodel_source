```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.signal import chirp, find_peaks, peak_widths
```

```python
df = pd.read_excel("input.xlsx")
df
```

```python
# set E as index
df = df.set_index("E")
```

```python
# X = df.columns
X = np.arange(-90, 100, 10)
```

```python
# normalize by mean
df.loc[:, 0:180] = df.loc[:,0:180].div(df.sum(axis=1), axis=0)
df
```

```python
# visualize some angle distributions
plt.scatter(X, df.iloc[0])
plt.scatter(X, df.iloc[5])
```

```python
def gauss(x, A, sigma):
    return A * np.exp(-(x - 0)**2/(2*(sigma)** 2))
```

```python
E05 = df.iloc[0]
E55 = df.iloc[5]
```

```python
# when do fitting all axis must be normalized
x = X/X.max()
popt05, pcov05 = curve_fit(gauss, x, E05)
popt55, pcov55 = curve_fit(gauss, x, E55)
```

```python
y05 = gauss(x, *popt05)
y55 = gauss(x, *popt55)
```

```python
# visualize fitting to E05
plt.scatter(X, E05, color="red")
plt.plot(X, y05)
```

```python
# visualize fitting to E55
plt.scatter(X, E55, color="red")
plt.plot(X, y55)
```

```python
# find peaks and their width
peaks05, _ = find_peaks(y05)
width05 = peak_widths(y05, peaks05, rel_height=0.5)

peaks55, _ = find_peaks(y55)
width55 = peak_widths(y55, peaks55, rel_height=0.5)
width05, width55
```

```python
# visualize fitting to E05 data
plt.scatter(X, E05, color="red")
plt.plot(X, y05, "*")
plt.plot(X[peaks05], y05[peaks05], "X", color="black")
plt.hlines(width05[1], width05[2]*10, width05[3]*10, color="C2")
```

```python
# visualize fitting to E55 data
plt.scatter(X, E55, color="red")
plt.plot(X, y55)
plt.plot(X[peaks55], y55[peaks55], "X", color="black")
plt.hlines(width05[1], width55[2]*10, width55[3]*10, color="C2")
```

```python
# now we need to apply the steps above to the all energies
# the prefered output is array of width
# approximation by gauss
yis = dict()
for i in range(len(df)):
   # df.iloc[i] is Ei or Xi in other words
    popti, _ = curve_fit(gauss, x, df.iloc[i])
    print(popti)
    yis[df.index[i]] = np.array(gauss(x, *popti))
```

```python
# target diameter in phits model is 5cm
pnw = dict()
for k,v in yis.items():
    peaks, _ = find_peaks(v)
    width = peak_widths(v, peaks, rel_height=0.5)
    pnw[k] = dict() 
    pnw[k]["peak"] = peaks
    pnw[k]["width"] = width
    print("E: ", k, "FWHM: ", width[0][0]*10, "degree", width[0][0]*10/180*10, "cm")
```

```python
# create subplots and plot all gaussians
fig, axs = plt.subplots(7, 2, figsize=(10,20))
row, col = 0,0
for i in range(len(yis)):
#     print(row, col)
#     print(yis[list(yis.keys())[i]])
    label = list(yis.keys())[i]
    f = yis[list(yis.keys())[i]]
    peak = pnw[list(pnw.keys())[i]]["peak"]
    width = pnw[list(pnw.keys())[i]]["width"]
    axs[row,col].bar(X, f, width=8)
    axs[row,col].plot(X, f, label=label, color="red")
    axs[row,col].scatter(X, df.iloc[i], color="green")
    axs[row,col].plot(X[peak], f[peak], "x")
    axs[row,col].hlines(width[1], width[2]*10, width[3]*10, color="C2")
    axs[row,col].legend()
    if i == 6:
        row -= 6
        col = 1
    else:
        row += 1
```
